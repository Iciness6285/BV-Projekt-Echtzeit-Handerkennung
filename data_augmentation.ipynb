```python
#import libraries
import pandas as pd
import numpy as np
import re
```

```python
#load data
file_path = 'train_sortiert_new_format.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
#create empty time laps dataframe
df_tl = pd.DataFrame(columns=df.columns)
df_tl.head()
```

```python
#fill in first columns
df_tl[["video_id","label_id","label"]]=df[["video_id","label_id","label"]]
```

```python
# fill in the data from every second frame of df
pattern = r'^F\d*[13579].*'
counter = 3
for col in df.columns:
    if re.match(pattern,col):
        df_tl.iloc[:,counter]=df.loc[:,col]
        counter += 1
```

```python
#fill other columns with 0
while counter < 2334:
    df_tl.iloc[:,counter]= 0
    counter += 1
```

```python
df_tl.shape
```

```python
#concatenate both dataframes
df_new = pd.concat([df,df_tl])
df_new.shape
```
